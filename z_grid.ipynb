```python

import numpy as np
import matplotlib.pyplot as plt
```

```python
dz0 = 3.2e-4

N_pix = 500

idx_z = np.arange(N_pix)

dz = dz0 * (1 + (5*idx_z/N_pix)**2.5)

z_grid = np.cumsum(dz)

plt.plot(z_grid)
```
